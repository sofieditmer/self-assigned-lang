__Data wrangling__

```python
filepath = os.path.join("..", "data", "wiki_movie_plots.csv")
movie_data = pd.read_csv(filepath)
```

```python
movie_data = movie_data.loc[:, ("Title", "Genre", "Plot")]
```

```python
# Filter out movies with unknown genre
movie_data = movie_data[~movie_data['Genre'].isin(['unknown'])]
```

```python
# Making new column for clean genre discriptioned
movie_data["Genre_clean"] = movie_data["Genre"] 
```

```python
# Combine genres "sci-fi" with "science fiction" 
movie_data["Genre_clean"].replace({"sci-fi": "science fiction"}, inplace=True)

# Combine "romantic comedy" with "romance"
movie_data["Genre_clean"].replace({"romantic comedy": "romance"}, inplace=True)
```

```python
# Choosing genres based on their frequency
chosen_genres = movie_data["Genre_clean"].value_counts().reset_index(name="count").query("count > 1000")["index"].tolist()
movie_data = movie_data[movie_data["Genre_clean"].isin(chosen_genres)].reset_index(drop=True)
```

```python
# Sample roughly equal number of movie plots from different genres (to reduce class imbalance issues)
movie_data_balanced = movie_data.groupby("Genre_clean").sample(1000).reset_index(drop=True)
```

```python
movie_data_balanced = movie_data_balanced.loc[:, ("Title", "Plot", "Genre_clean")]
```

__Create train and validation data__

```python
sentences = movie_data_balanced['Plot'].values # taking sentences
labels = movie_data_balanced['Genre_clean'].values # taking labels
```

```python
# train and test split using sklearn
X_train, X_test, y_train, y_test = train_test_split(sentences, 
                                                    labels, 
                                                    test_size=0.25, 
                                                    random_state=42)
```

__Binarize labels__

```python
lb = LabelBinarizer()
y_train_binarized = lb.fit_transform(y_train)
y_test_binarized = lb.fit_transform(y_test)
```

__Create word embeddings__

```python
# initialize tokenizer
tokenizer = Tokenizer(num_words=5000) # vocabulary = 5000 words
# fit to raw training data
tokenizer.fit_on_texts(X_train)

# tokenized training and test data
# Hence, we use the tokenizer to create sequences of tokens
X_train_tokens = tokenizer.texts_to_sequences(X_train)
X_test_tokens = tokenizer.texts_to_sequences(X_test)

# overall vocabulary size
vocab_size = len(tokenizer.word_index) + 1  # Adding 1 because of reserved 0 index
```

__Add padding__

```python
# max length for a doc
maxlen = max(len(max(X_train_tokens)), len(max(X_test_tokens, key=len))) 

# pad training data to maxlen
X_train_pad = pad_sequences(X_train_tokens, 
                            padding='post', # post = puts the pads at end of the sequence. Sequences can be padded "pre" or "post"
                            maxlen=maxlen)
# pad testing data to maxlen
X_test_pad = pad_sequences(X_test_tokens, 
                           padding='post', 
                           maxlen=maxlen)
```

__Create embedding matrix__

```python
def create_embedding_matrix(filepath, word_index, embedding_dim):
    """ 
    A helper function to read in saved GloVe embeddings and create an embedding matrix
    
    filepath: path to GloVe embedding
    word_index: indices from keras Tokenizer
    embedding_dim: dimensions of keras embedding layer
    """
    vocab_size = len(word_index) + 1  # Adding again 1 because of reserved 0 index
    embedding_matrix = np.zeros((vocab_size, embedding_dim))

    with open(filepath) as f:
        for line in f:
            word, *vector = line.split()
            if word in word_index:
                idx = word_index[word] 
                embedding_matrix[idx] = np.array(
                    vector, dtype=np.float32)[:embedding_dim]

    return embedding_matrix
```

```python
embedding_matrix = create_embedding_matrix(os.path.join("..", "glove", "glove.6B.100d.txt"),
                                           tokenizer.word_index, # keeping only the words that appear in our data
                                           embedding_dim=100)
```

__Define and train model__

```python
# Clearing sessions (deleting already trained models)
tf.keras.backend.clear_session()
```

```python
embedding_dim = 100
l2 = L2(0.0001)

# New model
model = Sequential()

# Embedding -> CONV+ReLU -> MaxPool -> FC+ReLU -> Out
model.add(Embedding(vocab_size,                  # vocab size from Tokenizer()
                    embedding_dim,               # embedding input layer size
                    weights=[embedding_matrix],  # pretrained embeddings
                    input_length=maxlen,         # maxlen of padded doc
                    trainable=False))            # trainable embeddings

model.add(Conv1D(128, 5, 
                activation='relu',
                kernel_regularizer=l2))          # L2 regularization 

model.add(GlobalMaxPool1D())

model.add(Dense(10, activation='relu', kernel_regularizer=l2))

model.add(Dense(5, activation='softmax'))

# compile
model.compile(loss='categorical_crossentropy',
              optimizer="adam",
              metrics=['accuracy'])

# print summary
model.summary()
```

__Train model__

```python
def plot_history(H, epochs):
    """
    Utility function for plotting model history using matplotlib
    
    H: model history 
    epochs: number of epochs for which the model was trained
    """
    plt.style.use("fivethirtyeight")
    plt.figure()
    plt.plot(np.arange(0, epochs), H.history["loss"], label="train_loss")
    plt.plot(np.arange(0, epochs), H.history["val_loss"], label="val_loss")
    plt.plot(np.arange(0, epochs), H.history["accuracy"], label="train_acc")
    plt.plot(np.arange(0, epochs), H.history["val_accuracy"], label="val_acc")
    plt.title("Training Loss and Accuracy")
    plt.xlabel("Epoch #")
    plt.ylabel("Loss/Accuracy")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
# Train model
history = model.fit(X_train_pad, y_train_binarized,
                    epochs=10,
                    verbose=True,
                    validation_data=(X_test_pad, y_test_binarized),
                    batch_size=20)

# Predictions
predictions = model.predict(X_test_pad, batch_size=20)
    
# Classification report
classification_metrics = classification_report(y_test_binarized.argmax(axis=1),
                                               predictions.argmax(axis=1),
                                               target_names=set(labels))
        
# Training accuracy
train_loss, train_accuracy = model.evaluate(X_train_pad, y_train_binarized, verbose=False)
        
# Validation accuracy
test_loss, test_accuracy = model.evaluate(X_test_pad, y_test_binarized, verbose=False)
        
# plot
plot_history(history, epochs = 10)
```

__Plot__

```python
unique_labels = list(set(labels))

plt.figure(figsize=(20, 10))
p1 = plt.bar(unique_labels, height = train_accuracy)
p2 = plt.bar(unique_labels, height = test_accuracy)
plt.title("Movies Genre Classification Accuracy")
plt.ylabel('Accuracy', fontsize=20)
plt.xlabel('Movie Genres', fontsize=20)
plt.legend((p1[0], p2[0]), ('Training accuracy', 'Test Accuracy'))
plt.show()
```
